```python
import matplotlib as plt
import numpy as np
import pandas as pd
from glob import glob as gg
```

```python
data_list= ['boost_dotp_k40.txt', 'boost_dotp_rtx.txt', 'cuda_dotp_k40.txt', 'cuda_dotp_rtx.txt', 'thrust_dotp_k40.txt', 'thrust_dotp_rtx.txt', 'vexcl_dotp_k40.txt', 'vexcl_dotp_rtx.txt', 'viennacl_dotp_k40.txt', 'viennacl_dotp_rtx.txt']

data = pd.DataFrame()
data['N']=pd.read_csv('data/boost_dotp_k40.txt', delimiter=" ")['N']

for i in data_list:
    datai=pd.read_csv('data/'+i, delimiter=" ")
    del datai['N']
    data[i[:-4]]=datai

data
```
